```python
import matplotlib.pyplot as plt
import os
import numpy as np
import re
```

## Help functions

```python
def load_vector(filename):
    if not os.path.exists(filename):
        print(f"ERROR: File '{filename}' not found")
    else:
        with open(filename, "r") as file:
            # Read the entire file as a string
            file_contents = file.read()

        # Split and save values to list
        lines = file_contents.split("\n")
        list = [float(item) for item in lines]
        return np.array(list)
```

## Euler integrator plots

### $\Delta t = 0.01$

```python
filename_first = r"data/ex_1/first_euler_dt=0.010000_"
filename_second = r"./data/ex_1/second_euler_dt=0.010000_"
filename_analytical = r"./data/ex_1/analytical_hc_dt=0.010000_"

xvals_first_t1 = load_vector(filename_first + "x_vals")
pvals_first_t1 = load_vector(filename_first + "p_vals")
xvals_second_t1 = load_vector(filename_second + "x_vals")
pvals_second_t1 = load_vector(filename_second + "p_vals")
xvals_analytical = load_vector(filename_analytical + "x_vals")
pvals_analytical = load_vector(filename_analytical + "p_vals")
delta_t = 0.01
T = 10
ts = np.linspace(0, 10, num=int(T/delta_t))
```

## 1.d Plot of Hamiltonian and H' for the symplectic integrator

```python
H = lambda x,p,k,m: p**2/(2*m) + k*x**2/2
H_shadow = lambda x,p,k,m,dt: H(x,p,k,m) - p*x*dt/2
x_0 = 1
p_0 = 0
m = 1
k = 1
```

$\Delta t = 0.01$

```python

def plot_Hs_t1():
    dt = 0.01
    ts = np.linspace(0,T, num=int(T/dt))
    H_analytical = np.zeros(len(ts))
    H_first = np.zeros(len(ts))
    H_second = np.zeros(len(ts))
    H_shadow_second = np.zeros(len(ts))

    for i in range(len(ts)):
        H_analytical[i] = H(xvals_analytical[i], pvals_analytical[i],k,m)
        H_first[i] = H(xvals_first_t1[i], pvals_first_t1[i],k,m)
        H_second[i] = H(xvals_second_t1[i], pvals_second_t1[i],k,m)
        H_shadow_second[i] = H_shadow(xvals_second_t1[i], pvals_second_t1[i],k,m,dt)

    fig,ax = plt.subplots()

    # ax.plot(ts, H_analytical, label="analytical")
    ax.plot(ts, H_first, label="H first integrator")
    ax.plot(ts, H_second, label="H second")
    ax.plot(ts, H_shadow_second, label="H'second shadow")
    ax.set_title("H's for dt=0.01")
    ax.set_xlabel("t")

    ax.legend()
plot_Hs_t1()
```

The first integrator has a longer time drift in the energy estimate than the second one. One can also note that the shadow hamiltonian for the symplectic integrator stays constant, as mentioned in task 1.a

$\Delta t = 0.001$

```python
def plot_Hs_t2():
    dt = 0.001
    ts = np.linspace(0,T, num=int(T/dt))
    H_analytical = np.zeros(len(ts))
    H_first = np.zeros(len(ts))
    H_second = np.zeros(len(ts))
    H_shadow_second = np.zeros(len(ts))

    for i in range(len(ts)):
        H_first[i] = H(xvals_first_t2[i], pvals_first_t2[i],k,m)
        H_second[i] = H(xvals_second_t2[i], pvals_second_t2[i],k,m)
        H_shadow_second[i] = H_shadow(xvals_second_t2[i], pvals_second_t2[i],k,m,dt)

    fig,ax = plt.subplots()

    ax.plot(ts, H_first, label="H first integrator")
    ax.plot(ts, H_second, label="H second")
    ax.plot(ts, H_shadow_second, label="H'second shadow")
    ax.set_title("H's for dt=0.001")
    ax.set_xlabel("t")

    ax.legend()
plot_Hs_t2()
```

Its the same case as for $\Delta t = 0.01$, but with a lower energy scale
