```python
import matplotlib.pyplot as plt
import os
import numpy as np
import re
```

## Help functions

```python
def load_vector(filename):
    if not os.path.exists(filename):
        print(f"ERROR: File '{filename}' not found")
    else:
        with open(filename, "r") as file:
            # Read the entire file as a string
            file_contents = file.read()

        # Split and save values to list
        lines = file_contents.split("\n")
        list = [float(item) for item in lines]
        return np.array(list)
```

## Euler integrator plots

### $\Delta t = 0.01$

```python
filename_first = r"data/ex_1/first_euler_dt=0.010000_"
filename_second = r"./data/ex_1/second_euler_dt=0.010000_"
filename_analytical = r"./data/ex_1/analytical_hc_dt=0.010000_"

xvals_first = load_vector(filename_first + "x_vals")
pvals_first = load_vector(filename_first + "p_vals")
xvals_second = load_vector(filename_second + "x_vals")
pvals_second = load_vector(filename_second + "p_vals")
xvals_analytical = load_vector(filename_analytical + "x_vals")
pvals_analytical = load_vector(filename_analytical + "p_vals")
delta_t = 0.01
T = 10
ts = np.linspace(0, 10, num=int(T/delta_t))
```

```python
fig, ax = plt.subplots(2,1)

#X plot
ax[0].plot(ts, xvals_first, label="First Euler")
ax[0].plot(ts, xvals_second, label="Second Euler")
ax[0].plot(ts, xvals_analytical, label="Analytical HC")

ax[0].axhline(y=1, color='r', linestyle='--', label='y=1')
ax[0].axhline(y=-1, color='r', linestyle='--', label='y=-1')
ax[0].axvline(x=np.pi, color='r', linestyle='--', label=r'x=$\pi$')

ax[0].legend()
ax[0].set_title(r"Comparision of Euler Integrators for $\Delta t=0.01$")
ax[0].set_ylabel("X(t)")

#P plot
ax[1].plot(ts, pvals_first, label="First Euler")
ax[1].plot(ts, pvals_second, label="Second Euler")
ax[1].plot(ts, pvals_analytical, label="Analytical HC")

ax[1].axhline(y=1, color='r', linestyle='--', label='y=1')
ax[1].axhline(y=-1, color='r', linestyle='--', label='y=-1')
ax[1].axvline(x=np.pi, color='r', linestyle='--', label=r'x=$\pi$')

ax[1].legend()
# ax[1].set_title(r"Comparision of Euler Integrators for $\Delta t=0.01$")
ax[1].set_ylabel("P(t)")
```

### Error

```python
fig, ax = plt.subplots(2,1)

ax[0].plot(ts, abs(np.subtract(xvals_first,xvals_analytical)), label="First Euler")
ax[0].plot(ts, abs(np.subtract(xvals_second,xvals_analytical)), label="Second Euler")



ax[0].legend()
ax[0].set_title(r"Error of Euler Integrators for $\Delta t=0.01$")
ax[0].set_ylabel("X(t)")

#P plot
ax[1].plot(ts, abs(np.subtract(pvals_first,pvals_analytical)), label="First Euler")
ax[1].plot(ts, abs(np.subtract(pvals_second,pvals_analytical)), label="Second Euler")


ax[1].legend()
ax[1].set_ylabel("P(t)")
```

## 1.d Plot of Hamiltonian and H' for the symplectic integrator

```python
H = lambda x,p,k,m: p**2/(2*m) + k*x**2/2
H_shadow = lambda x,p,k,m,dt: H(x,p,k,m) - p*x*dt/2
x_0 = 1
p_0 = 0
m = 1
k = 1
```

$\Delta t = 0.01$

```python
dt = 0.01
ts = np.linspace(0,T, num=int(T/dt))
H_analytical = np.zeros(len(ts))
H_first = np.zeros(len(ts))
H_second = np.zeros(len(ts))
H_shadow_second = np.zeros(len(ts))
for i in range(len(ts)):
    H_analytical[i] = H(xvals_analytical[i], pvals_analytical[i],k,m)
    H_first[i] = H(xvals_first[i], pvals_first[i],k,m)
    H_second[i] = H(xvals_second[i], pvals_second[i],k,m)
    H_shadow_second[i] = H_shadow(xvals_second[i], pvals_second[i],k,m,dt)

fig,ax = plt.subplots()

ax.plot(ts, H_analytical, label="analytical")
ax.plot(ts, H_first, label="first")
ax.plot(ts, H_second, label="second")
ax.plot(ts, H_shadow_second, label="second shadow")

ax.legend()
```
