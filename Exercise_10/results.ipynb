## Help functions

```python
def load_vector(filename):
    if not os.path.exists(filename):
        print(f"ERROR: File '{filename}' not found")
    else:
        with open(filename, "r") as file:
            # Read the entire file as a string
            file_contents = file.read()

        # Split and save values to list
        lines = file_contents.split("\n")
        list = [float(item) for item in lines]
        return np.array(list)
```

## 1.d Plot of Hamiltonian and H' for the symplectic integrator

```python
H = lambda x,p,k,m: p**2/(2*m) + k*x**2/2
H_shadow = lambda x,p,k,m,dt: H(x,p,k,m) - p*x*dt/2
x_0 = 1
p_0 = 0
m = 1
k = 1
```

$\Delta t = 0.01$

```python

def plot_Hs_t1():
    dt = 0.01
    ts = np.linspace(0,T, num=int(T/dt))
    H_analytical = np.zeros(len(ts))
    H_first = np.zeros(len(ts))
    H_second = np.zeros(len(ts))
    H_shadow_second = np.zeros(len(ts))

    for i in range(len(ts)):
        H_analytical[i] = H(xvals_analytical[i], pvals_analytical[i],k,m)
        H_first[i] = H(xvals_first_t1[i], pvals_first_t1[i],k,m)
        H_second[i] = H(xvals_second_t1[i], pvals_second_t1[i],k,m)
        H_shadow_second[i] = H_shadow(xvals_second_t1[i], pvals_second_t1[i],k,m,dt)

    fig,ax = plt.subplots()

    # ax.plot(ts, H_analytical, label="analytical")
    ax.plot(ts, H_first, label="H first integrator")
    ax.plot(ts, H_second, label="H second")
    ax.plot(ts, H_shadow_second, label="H'second shadow")
    ax.set_title("H's for dt=0.01")
    ax.set_xlabel("t")

    ax.legend()
plot_Hs_t1()
```

The first integrator has a longer time drift in the energy estimate than the second one. One can also note that the shadow hamiltonian for the symplectic integrator stays constant, as mentioned in task 1.a

$\Delta t = 0.001$

```python
def plot_Hs_t2():
    dt = 0.001
    ts = np.linspace(0,T, num=int(T/dt))
    H_analytical = np.zeros(len(ts))
    H_first = np.zeros(len(ts))
    H_second = np.zeros(len(ts))
    H_shadow_second = np.zeros(len(ts))

    for i in range(len(ts)):
        H_first[i] = H(xvals_first_t2[i], pvals_first_t2[i],k,m)
        H_second[i] = H(xvals_second_t2[i], pvals_second_t2[i],k,m)
        H_shadow_second[i] = H_shadow(xvals_second_t2[i], pvals_second_t2[i],k,m,dt)

    fig,ax = plt.subplots()

    ax.plot(ts, H_first, label="H first integrator")
    ax.plot(ts, H_second, label="H second")
    ax.plot(ts, H_shadow_second, label="H'second shadow")
    ax.set_title("H's for dt=0.001")
    ax.set_xlabel("t")

    ax.legend()
plot_Hs_t2()
```

Its the same case as for $\Delta t = 0.01$, but with a lower energy scale

### Energy conservation

```python
def total_mek_energy(rs, vs, k=10, m=1):
    U = 1/2*k*rs**2
    K = 1/2*m*vs**2
    return U+K
```

```python
# Load Beeman values
rs_b = np.loadtxt("./data/ex_2/beeman/0.100000/r.dat")
vs_b = np.loadtxt("./data/ex_2/beeman/0.100000/v.dat")
fs_b = np.loadtxt("./data/ex_2/beeman/0.100000/f.dat")
ts_b = np.loadtxt("./data/ex_2/beeman/0.100000/t.dat")

# Load Velocity Verlet values
xs_vv = np.loadtxt("./data/ex_2/vv/0.100000/x.dat")
vs_vv = np.loadtxt("./data/ex_2/vv/0.100000/v.dat")
fs_vv = np.loadtxt("./data/ex_2/vv/0.100000/f.dat")
ts_vv = np.loadtxt("./data/ex_2/vv/0.100000/t.dat")
```

```python
def print_values(rs,vs,fs,ts, model_name):
    print("=====================================")
    print("Model: {0}".format(model_name))
    print("initial values:")
    print("E: {0}".format(total_mek_energy(rs[0], vs[0])))
    print("f = {:.3f}".format(fs[0]))
    print("v = {:.3f}".format(vs[0]))
    print("r = {:.3f}".format(rs[0]))
    print("dt: {0}".format(ts[1]-ts[0]))
    print("Final values:")
    print("f = {:.3f}".format(fs[-1]))
    print("v = {:.3f}".format(vs[-1]))
    print("r = {:.3f}".format(rs[-1]))
    print("Total energy: {:.3f}".format(total_mek_energy(rs[-1], vs[-1])))
    print("Delta E: {:.3f}".format(total_mek_energy(rs[-1], vs[-1]) - total_mek_energy(rs[0], vs[0])))
    print("=====================================")

print_values(rs_b, vs_b, fs_b, ts_b, "Beeman")
# print_values(rs_v, vs_v, fs_v, ts_v, "Verlet")
print_values(xs_vv, vs_vv, fs_vv, ts_vv, "Velocity Verlet")
```

Energy harmonic oscillator:

$E = K + U$

$K_{start} = 0$

$ U_{start} = \int{f dx} = \frac{1}{2} kx_i^2$

with $k = 10$ and $x_i = 5$, $ E_{start} = 125$ for all models

#### Beeman final energy: $E_f = 44.362$

#### Velocity Verlet final energy: $E_f = 123.832$

##### Comments:
One can see that the Velocity Verlet is symplectic, as it (almost) keeps its energy conserved during the simulation, while the Beeman is much more unprecise, but can be improved by decreasing the timestep. Both integrators have an accuracy of second order ($O(\Delta t^2)$), but only the velocity verlet is symplectic of these two.

There could be other situations than a simple spring model were the Beeman algorithm could be more accurate than the Velocity Verlet, maybe a situation were the acceleration is more chaotic, since it uses the acceleration in a more detailed way than the Velocity Verlet

# Exercise 3 Harmonic Oscillator #2

```python
#IC
ts = np.linspace(0,10, num=1000)
dt = ts[1]-ts[0]
w = 2
p_0 = 0
q_0 = 1
qs = np.zeros(len(ts))
ps = np.zeros(len(ts))
qs[0] = q_0
ps[0] = p_0

#First step
q_now = q_0 + p_0*dt
p_now = p_0 + (-w**2*q_0)*dt
qs[1] = q_now
ps[1] = p_now
for i in range(2, len(ts)):
    q_now += p_now*dt
    p_now += (-w**2*q_now)*dt
    qs[i] = q_now
    ps[i] = p_now

#Plotting
ps = np.array(ps)
qs = np.array(qs)
plt.figure(figsize=(6,6))
plt.plot(ps, w*qs, label="analytical sol")
plt.xlabel("p")
plt.ylabel(r"$\omega q$")
plt.legend()
plt.show()
```
